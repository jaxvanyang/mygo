```python
import time
import random
import itertools

from go.game import Game
from agent.naive import RandomBot
from agent.helper import *
```

```python
%%prun -l 15 -s cumulative
# RandomBot vs. itself (with Zobrist optimization)

random.seed(25565)  # for reproducible output
size = 13
game = Game.new_game(size)
bot = RandomBot()

start_time = time.perf_counter()

for i in itertools.count(1):
    if game.is_over:
        break

    move = bot.select_move(game)
    print_move(game.next_color, move, i)
    # print(game.board)
    game.apply_move(move)

end_time = time.perf_counter()

print()
print(game.board)
print()
print(f"time: {end_time - start_time:.3f}s")
```
